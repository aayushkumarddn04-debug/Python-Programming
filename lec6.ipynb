create a class of animal and a sub class of pets 

```python
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age
        self.vaccinated = False 

    def vaccinate(self):
        if not self.vaccinated:
            self.vaccinated = True
            print(f"{self.name} has been vaccinated.")
        else:
            print(f"{self.name} is already vaccinated.")

    def info(self):
        print(f"Name: {self.name}")
        print(f"Species: {self.species}")
        print(f"Age: {self.age}")
        print(f"Vaccinated: {'Yes' if self.vaccinated else 'No'}")


class Pet(Animal):
    def __init__(self, name, species, age, owner):
        super().__init__(name, species, age)
        self.owner = owner

    def vaccinate_pet(self):
        print(f"Vaccinating pet owned by {self.owner}...")
        super().vaccinate()  

    def pet_info(self):
        self.info()
        print(f"Owner: {self.owner}")

```

```python
pet1 = Pet("Buddy", "Dog", 3, "Aayush")

pet1.pet_info()
print()
pet1.vaccinate_pet()
print()
pet1.pet_info()
```
