```python
def top_k(items, k):
    sorted_items = sorted(items, key=lambda x: x[1], reverse=True)
    return sorted_items[:k]
```

```python
items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2

print(top_k(items, k))
```

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] += 1
    return d
```

```python
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
print(count_by_category(pairs))
```

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2 
    return total - sum(nums)
```

```python
nums = [0,3,1]
missing_number(nums)
```

```python
def moving_average(stream,w):
    if w<=0:
        raise ValueError("window size must be greater than 0")
    if w>len(stream):
        raise ValueError("window size cannot be greater than any arr length")
    
    result = []

    for i in range (len(stream) -w + 1):
     window = stream[i:i+w]          
     avg = sum(window) / w        
     result.append(avg)
    
    return result
```

```python
def rle_decode(s):
    result = []
    i = 0
    n = len(s)

    while i < n:
    
        if not s[i].isdigit():
            raise ValueError("Invalid format: count expected")

        num = 0
        while i < n and s[i].isdigit():
            num = num * 10 + int(s[i])
            i += 1

        if i >= n:
            raise ValueError("Invalid format: character missing after count")

        
        if s[i] == '\\':
            i += 1
            if i >= n:
                raise ValueError("Invalid escape sequence")
            char = s[i]
        else:
            char = s[i]

        result.append(char * num)
        i += 1

    return ''.join(result)



print(rle_decode("3a2b1c"))  
print(rle_decode("2\\3"))    
```

```python
def longest_unique_substring(s):
    char_index = {}
    left = 0
    max_length = 0

    for right in range(len(s)):
        if s[right] in char_index and char_index[s[right]] >= left:
            left = char_index[s[right]] + 1

        char_index[s[right]] = right
        max_length = max(max_length, right - left + 1)

    return max_length



print(longest_unique_substring("abcabcbb")) 
```
