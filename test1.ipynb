## Mutable Data Types

Mutable objects allow in-place modification without changing the object identity.

```python
numbers = [10, 20, 30]
numbers[0] = 99
print(numbers)
```

## Memory Identity Example

Immutable type:

```python
x = 5
print(id(x))
x = x + 1
print(id(x))
```

Mutable type:

```python
a = [1, 2]
print(id(a))
a.append(3)
print(id(a))
```
